```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer,StandardScaler
from sklearn.metrics import accuracy_score ,precision_score,recall_score, f1_score
```

```python
df = pd.read_csv("/content/sample_data/heart.csv")
```

```python
df.drop_duplicates(inplace=True)
print("no. of duplicates: ",df.duplicated().sum())
```

```python
x=df.iloc[:,:13]
y=df.iloc[:,13]
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
b = np.ones((x.shape[0], 1))
x= np.hstack((x, b))
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,shuffle = True)
```

```python
#generating the weights+bias
weights=np.random.rand(14,14)
```

```python
# wtranspose x
w_transpose_x=np.dot(weights, x_train.T);
def sigmoid(x):
  return 1/(1 + np.e**-x)

y_pred=sigmoid(w_transpose_x)
print(y_pred.shape)
```

```python
def train(epochs,weights,x_train,y_pred,y_train,learning_rate):
  for i in range(epochs):
    for j in range(len(weights)):
      ys=(y_pred[j]-y_train)
      change=1/241*(learning_rate *np.dot(x_train.T,ys))
      weights[j]=weights[j]-change
  return weights

weights=train(200,weights,x_train,y_pred,y_train,0.1)
```

```python
def test(weights,x_test):
  predictions=[]
  y_pred=np.dot(weights,x_test.T)
  y_pred=sigmoid(y_pred)
  return y_pred

y_predict=test(weights,x_test)
for i in range(len(y_predict)):
  for j in range(len(y_predict[i])):
    if y_predict[i][j]<=0.5:
      y_predict[i][j]=0
    else:
      y_predict[i][j]=1

max_index=0
max_accuracy=-1
for k in range(len(y_predict)):
  if(accuracy_score(y_predict[i],y_test) > max_accuracy):
    max_accuracy=accuracy_score(y_predict[i],y_test)
    max_index=k

print("The accuracy is: ",accuracy_score(y_test,y_predict[max_index]))
print("The precision is: ",precision_score(y_test,y_predict[max_index]))
print("Recall is: ",recall_score(y_test,y_predict[max_index]))
print('F1-score:', f1_score(y_test,y_predict[max_index]))
```
